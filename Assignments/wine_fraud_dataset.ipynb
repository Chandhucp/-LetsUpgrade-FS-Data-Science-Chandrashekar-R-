```python
X = df.drop(['quality', 'type'], axis=1)
y = df['quality']
```

```python
X = df.drop(['quality', 'type'], axis=1)
y1 = df['type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train, X_test, y1_train, y1_test = train_test_split(X, y1, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python

from sklearn.svm import SVR, LinearSVR
```

```python
help(SVR)
```

```python
base_model = SVR()
```

```python
base_model.fit(scaled_X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=1)
```

```python
knn_model.fit(scaled_X_train, y_train)
```

```python
knn_model.fit(scaled_X_train, y1_train)
```

```python
y_pred = knn_model.predict(scaled_X_test)
```

```python
y1_pred = knn_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
accuracy_score(y1_test, y1_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
confusion_matrix(y1_test, y1_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(classification_report(y1_test, y1_pred))
```

```python
error_rate = []

for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(scaled_X_train, y_train)
    
    y_pred_test = knn_model.predict(scaled_X_test)
    test_error = 1 - accuracy_score(y_test, y_pred_test)
    error_rate.append(test_error)
```

```python
error_rate1 = []

for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(scaled_X_train, y1_train)
    
    y1_pred_test = knn_model.predict(scaled_X_test)
    test_error = 1 - accuracy_score(y1_test, y1_pred_test)
    error_rate1.append(test_error)
```

```python
plt.figure(figsize=(10,8))

plt.plot(range(1,30), error_rate, label='Test Error')
plt.legend()
plt.xlabel('K values')
plt.ylabel('Error rates')
```

```python
plt.figure(figsize=(10,8))

plt.plot(range(1,30), error_rate1, label='Test Error')
plt.legend()
plt.xlabel('K values')
plt.ylabel('Error rates')
```

```python
scaler = StandardScaler()
```

```python
knn = KNeighborsClassifier()
```

```python
knn.get_params().keys()
```

```python

operations = [('scalar', scaler), ('knn', knn)]
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(operations)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
k_values = list(range(1,20))
```

```python
param_grid = {'knn__n_neighbors': k_values}
```

```python
full_cv_classifier = GridSearchCV(pipe,param_grid, cv=5, scoring='accuracy')
```

```python
full_cv_classifier.fit(X_train, y_train)
```

```python
full_cv_classifier.fit(X_train, y1_train)
```

```python
full_cv_classifier.best_estimator_.get_params()
```

```python
full_cv_classifier.cv_results_['mean_test_score']
```

```python

scaler = StandardScaler()
knn14 = KNeighborsClassifier(n_neighbors=14)
operations = [('scalar', scaler), ('knn14', knn14)]
```

```python
pipe = Pipeline(operations)
```

```python
pipe.fit(X_train, y_train)
```

```python
pipe.fit(X_train, y1_train)
```

```python
pipe_pred = pipe.predict(X_test)
```

```python
print(classification_report(y_test, pipe_pred))
```

```python
print(classification_report(y1_test, pipe_pred))
```
